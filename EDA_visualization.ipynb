For EDA visualization techniques, we use the iris dataset and the dataset containing some BUX values

```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

iris = pd.read_csv('/content/drive/MyDrive/datasets/iris (1).txt',
                   sep=';',
                   decimal=',',
                   names=['slength', 'swidth', 'plength', 'pwidth', 'iclass'])

bux = pd.read_csv('/content/drive/MyDrive/datasets/BUX_1ho (1).txt',
                   sep='\t',
                   decimal=',',
                   header=0)
bux['datum'] = pd.to_datetime(bux['datum'], format='%Y.%m.%d')
```

```python
bux.head()
```

**Line chart**

The evolution of the BUX index in the scanned time interval can be represented, for example, as follows: the 2 diagrams are drawn with the plot() function, then gcf() returns the current Figure, on which the autofmt_xdate() method sets the direction and location of the x-axis labels in a way that makes it more readable.

```python
plt.plot(bux['datum'], bux['kezdo'], marker='o')
plt.plot(bux['datum'], bux['zaro'], marker='o')
plt.gcf().autofmt_xdate()
plt.show()
```

```python
bux.plot(x='datum', marker='o')
plt.show()
```

**Dot-to-dot diagram**

Plotting the width of the sepal as a function of the length of the sepal!

```python
iris.plot(kind='scatter', x='slength', y='swidth')
plt.show()
```

```python
names = set(iris['iclass'])
print(names)
for name in names:
    cond = iris['iclass'] == name
    plt.plot(iris.slength[cond], iris.swidth[cond],
             linestyle='none', marker='o', label=name)
plt.legend(numpoints=1)
plt.show()
```

**Point-to-point**

With the point-to-point matrix diagram, we can create the point-to-point diagram for every pair of attributes with continuous values.

```python
pd.plotting.scatter_matrix(iris)
plt.show()
```

**Paralell-axes**

```python
pd.plotting.parallel_coordinates(iris, 'iclass')
plt.show()
```

**+1**

```python
names = set(iris['iclass'])
fig1 = plt.figure()

ax1 = fig1.add_subplot(211)  # two rows 1 column first plot
for name in names:
    cond = iris['iclass'] == name
    ax1.scatter(iris.slength[cond], iris.swidth[cond], s=10, marker='s', label=name)
ax1.legend()

ax2 = fig1.add_subplot(212)
for name in names:
    cond = iris['iclass'] == name
    ax2.scatter(iris.plength[cond], iris.pwidth[cond], s=10, marker='s', label=name)
ax2.legend()
plt.show()
```
