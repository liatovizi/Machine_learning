```python
# Function to Create Line Figure to Compare Before and After Scalling
def createLineFigureTrain(list):
    plt.figure(figsize=(10,8))
    for i in list:
        sns.kdeplot(X_train[i])
    title = 'Before Scalling'
    plt.title(title)
    plt.show()
def createLineFigureNormalized(list):
    plt.figure(figsize=(10,8))
    for i in list:
        sns.kdeplot(normalized_x_train[i])
    title = 'After Scalling'
    plt.title(title)
    plt.show()

# Function to Evaluate and Store the Evaluation Metric Score
def metricScore(actual,predicted,name):
    mae = mean_absolute_error(actual,predicted)
    mse = mean_squared_error(actual,predicted)
    mape = mean_absolute_percentage_error(actual,predicted)
    r2 = r2_score(actual,predicted)
    resultStorage[name] = [mae,mse,mape,r2]
    print('Mean Absolute Erorr: ',mae)
    print('Mean Squared Erorr: ',mse)
    print('Mean Absolute Percentage Erorr: ',mape)
    print('R Squared Erorr: ',r2)

# Define Function to Make Predictions, Evaluate the Model and Plot the Regression Line
def evaluateModel(x_test = None, y_test = None, model = None, name='...', color = 'maroon'):
    predicted = model.predict(x_test)
    actual = y_test
    metricScore(actual,predicted,name)
    plt.figure(figsize=(12,8))
    plt.scatter(actual,predicted, c = color)
    mxp, mxa, mnp, mna = max(predicted), max(actual), np.min(predicted), np.min(actual)
    p1 = mxp if mxp>mxa else mxa
    p2 = mnp if mnp<mna else mna
    plt.plot([p1,p1],[p1,p2],'#000066')
    plt.xlabel('True Values')
    plt.ylabel('Predictions')
    plt.title(name)
    plt.axis('equal')
    plt.show()
```

```python
# Split the dataframe_variable
X = df5.drop(['6th Stage'],axis=1)
y = df5['6th Stage']

# Make Train and Test Variable
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 1)

# Make normalized dataframe
scaler = MinMaxScaler()
normalized_x_train = pd.DataFrame(scaler.fit_transform(X_train),columns = X_train.columns)
normalized_x_train.head()
normalized_x_test = pd.DataFrame(scaler.fit_transform(X_test),columns = X_test.columns)
```

```python
# Create Variable to Store Result
resultStorage = {}

# Linear Regression
modelLR = LinearRegression()
modelLR.fit(normalized_x_train,y_train)
evaluateModel(x_test = normalized_x_test, y_test = y_test, model = modelLR, name = 'Linear Regression')

# Decision Tree
modelDR = DecisionTreeRegressor(random_state = 1)
modelDR.fit(normalized_x_train,y_train)
evaluateModel(x_test = normalized_x_test, y_test = y_test, model = modelDR, name = 'Decision Tree', color = 'lightblue')

# Random Forest Regressor
modelRF = RandomForestRegressor(n_estimators = 100, verbose =4)
modelRF.fit(normalized_x_train,y_train)
evaluateModel(x_test = normalized_x_test, y_test = y_test, model = modelRF, name = 'Random Forest', color = 'Purple')

# Gaussian Naive Bayes
ModelGNB = GaussianNB()
ModelGNB.fit(normalized_x_train,y_train)
evaluateModel(x_test = normalized_x_test, y_test = y_test, model = ModelGNB, name = 'Naive Bayes', color = 'indigo')
```
